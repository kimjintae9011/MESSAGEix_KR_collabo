```python
# Importing required software packages
import pandas as pd
import os
import ixmp
import message_ix
from plotter import plotter
```

```python
#%% Setup IXMP
mp = ixmp.Platform()
scenlist = mp.scenario_list()
mp.open_db()
```

```python
#%% clonning new scenario
model = 'MESSAGEix_KR'
scen = 'baseline'
base = message_ix.Scenario(mp, model=model, scenario = scen, version=1)
scenario = base.clone(model, 'Demand_management', 'uing LP', keep_solution=False)
scenario.check_out()
```

```python
KR_demand = scenario.par('demand')
```

```python
KR_demand 
```

```python
# demand reduction 
def demand(row) :
    if row['year'] == 2015:
        return row['value']*1
    if row['year'] == 2020:
        return row['value']*0.90
    if row['year'] == 2025:
        return row['value']*0.90
    if row['year'] == 2030:
        return row['value']*0.90
    if row['year'] == 2035:
        return row['value']*0.90
    if row['year'] == 2040:
        return row['value']*0.90
    if row['year'] == 2045:
        return row['value']*0.90
    if row['year'] == 2050:
        return row['value']*0.90 
    if row['year'] == 2055:
        return row['value']*0.90
    if row['year'] == 2060:
        return row['value']*0.90
    if row['year'] == 2070:
        return row['value']*0.90
    if row['year'] == 2080:
        return row['value']*0.90
    if row['year'] == 2090:
        return row['value']*0.90
    if row['year'] == 2100:
        return row['value']*0.90     
```

```python
KR_demand ['value']= KR_demand .apply(demand, axis='columns')
```

```python
KR_demand
# KR_demand.to_excel('demand.xlsx')
```

```python
scenario.add_par('demand',KR_demand )
```

```python
scenario.commit(comment="reducing demand")
```

```python
scenario.solve(solve_options={'lpmethod': '4'})
```

```python
emission = scenario.var('EMISS')
```
